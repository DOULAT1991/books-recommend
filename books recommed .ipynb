```python
## 3 DATASETS
ratings=pd.read_csv('BX-Book-Ratings.csv',sep=";",error_bad_lines=False,encoding='latin-1')
```

```python
ratings.head()
```

```python
ratings.rename(columns={'User-ID':'user_id','ISBN':'isbn','Book-Rating':'book_rating'},inplace=True)
```

```python
ratings.head()
```

```python
x=ratings['user_id'].value_counts()>200
```

```python
x[x].shape
```

```python
y=x[x].index
```

```python
y
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings.shape
```

```python
ratings.head()
```

```python
ratings.info()
```

```python
ratings_with_book=ratings.merge(books, on ='isbn')
```

```python
ratings_with_book
```

```python
ratings_with_book.shape
```

```python
number_rating=ratings_with_book.groupby('title')['book_rating'].count().reset_index()
```

```python
number_rating.rename(columns={'book_rating':'no_of_ratings'},inplace=True)
```

```python
number_rating
```

```python
final_rating=ratings_with_book.merge(number_rating, on='title')
```

```python
final_rating
```

```python
final_rating.shape
```

```python
final_rating=final_rating[final_rating['no_of_ratings']>=50]
```

```python
final_rating.shape
```

```python
## THERE ARE ALSO SOME DUPLICATE VALUE
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='book_rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(algorithm='brute')
model.fit(book_sparse)
```

#### NOW TEST MODEL

```python
distances,suggestions=model.kneighbors(book_pivot.iloc[54,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distances
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
    print(book_pivot.index[suggestions[i]])
```

```python
book_pivot.index[54]
```

```python
np.where(book_pivot.index=='Animal Farm')[0][0]
```

```python
def recommend_book(book_name):
    book_id=np.where(book_pivot.index==book_name)[0][0]
    distances,suggestions=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)
    for i in range(len(suggestions)):
            print(book_pivot.index[suggestions[i]])
```

```python
recommend_book('Animal Farm')
```
